## pair density function

```python
Utot_list = [0.0, 0.1, 0.5, 2.5]
gamma_list = [-2.82*10**(-15), 0.0264, 0.506, 2.43]

results = np.load('histogram_results.npy', allow_pickle=True)

colours = ['#00065c', 'indianred', 'cornflowerblue', 'orange']

fig, ax = plt.subplots(1,2, figsize=(8,4))

ax[0].plot(results[0,0], results[0,3], '--', color = 'k', label=f'Analytical Solution')
for i, (radii, pcf_dx, pcf_dp, exact_sol) in enumerate(results):
    colour = colours[i % len(colours)]
    radii = np.array(radii)
    pcf_dx = np.array(pcf_dx)
    exact_sol = np.array(exact_sol)
    Utot = Utot_list[i]
    ax[0].loglog(radii, pcf_dx - 1, color=colour, label=f'$U_{{\\tau}}/2 = {Utot}$')
    ax[0].plot(radii, exact_sol, '--', color = colour)

ax[0].set_xlabel(r'$r / \Delta$')
ax[0].set_ylabel(r'$g(r, t) - 1$')
ax[0].set_ylim(10**(-1), 10**13)
#ax[0].title('Pair Correlation Function (log Scale)')
ax[0].legend(fontsize='7', frameon=False)

ax[0].axline((0.5,0.8), (0.8, 0.5), linewidth=2, color='grey', transform=ax[0].transAxes, alpha=0.5)
ax[0].text(0.7, 0.65,'-2',transform=ax[0].transAxes,  fontdict=dict(fontsize=10))
ax[0].text(-0.25, 1,'a',transform=ax[0].transAxes,  fontdict=dict(fontsize=12))

for i, (radii, pcf_dx, pcf_dp, exact_sol) in enumerate(results):
    colour = colours[i % len(colours)]
    radii = np.array(radii)
    pcf_dx = (np.array(pcf_dx) - 1) / 10**9
    exact_sol = np.array(exact_sol) / 10**9
    ax[1].plot(radii, pcf_dx , color = colour, label=f'Numerical Results for U={Utot_list[i % len(colours)]}')
    ax[1].plot(radii, exact_sol, '--', color = colour, label=f'Analytical Solution for U={Utot_list[i % len(colours)]}')
ax[1].set_xlim(0, 50)
ax[1].set_ylim(0, 400)
ax[1].set_xlabel(r'$r / \Delta$')
ax[1].set_ylabel(r'$g(r, t) - 1 (\times 10^{9})$')
#ax[1].set_title('Pair Correlation Function (Linear Scale)')
#ax[1].legend()
ax[1].text(-0.18, 1, 'b', transform=ax[1].transAxes, fontdict=dict(fontsize=12))

plt.tight_layout()
plt.savefig('figures/histo.pdf', bbox_inches='tight')
```
